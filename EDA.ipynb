# Approach-3 Filling Mean, Median, or Mode in place of Null values

```python

# When to replace Null with mean:
    # 1. data must be numerical
    # 2. data must be normally distributed
    # 3. Data should not have outliers

# When to replace Null with median:
    # 1. When we have Outliers
    # 2. When the data is skiewed

# When to replace Null with mode:
    # 1. When data is categorical/discrete
```

```python
# Sample data 

Netflix_episodes_finish_per_day = [8, 5, np.nan, 0, 7, 5, 0, np.nan, np.nan, 100]

df5 = pd.DataFrame(Netflix_episodes_finish_per_day, columns = ["Number"])

mean_value = df5.mean()
print(mean_value)

df5 = df5.fillna(mean_value)
df5.isnull().sum()
df5
```

```python
# Replace Null with Median

Netflix_episodes_finish_per_day = [8, 5, np.nan, 0, 7, 5, 0, np.nan, np.nan, 100]

df5 = pd.DataFrame(Netflix_episodes_finish_per_day, columns = ["Number"])

median_value = df5.median()
print(median_value)

df5 = df5.fillna(median_value)
df5.isnull().sum()
df5
```

# Appraoch-5: Interpolate(): 

```python
# estimate missing data/ null values in a continuous series
# Types of interpolate functions : 
    # 1.linear : if data is having linear relationship
                 # it cant handle the starting null values
                # linear will put the same last valid value in the last nan values
    # 2.polynomial : if you have polynomial relationship in data
                     # polynomial donot insert the last nan values 
                    # polynomial also cannot handle the starting nan values
    # 3.spline : smooth curve
                 # spline canot fill the starting null value
                 # spline handles the last nan values
    # 4. time : when we have time-series data
number_gf=[np.nan,1,2,3,4,5,np.nan,np.nan,8,9,np.nan,np.nan]
df5=pd.DataFrame(number_gf)
df5
```

```python
df5.interpolate(method="linear",inplace=True)
df5
```

```python
number_gf=[np.nan,10,20,30,40,50,np.nan,np.nan,8,9,np.nan,np.nan]
df5=pd.DataFrame(number_gf)
df5.interpolate(method="polynomial",order=2,inplace=True)
df5
```

```python
number_gf=[np.nan,10,20,30,40,50,np.nan,np.nan,8,9,np.nan,np.nan]
df5=pd.DataFrame(number_gf)
df5.interpolate(method="spline",order=2,inplace=True)
df5
```

```python
# finding which records are actually outliers 
Q1=df.age.quantile(.25)
print(Q1)
Q3=df.age.quantile(.75)
print(Q3)
IQR=Q3-Q1
print(IQR)

lower_bound=Q1-1.5*IQR
print(lower_bound)
upper_bound= Q3+1.5*IQR
print(upper_bound)
```

```python
# outliers : which are less than the lowerbound and greater than upperbound
outliers=df[(df.age<lower_bound)|(df.age>upper_bound)]
outliers
```

```python
df1 = df.copy()
# Fin the records where age > lower bound & age < upper bound
df1 = df1[(df1.age >= lower_bound) & (df1.age <= upper_bound)]
print(df1.shape)
print(df.shape)
```

# EDA-6: Identify/Handling of Outliers using z-score

```python
# Def: How many standard deviation, data point is away from the mean.

# Here we are assuming that data is normal distributed.
# We should apply z-score method to find the outliers if data is normally distributed
# if data point (age) z-score > 3 Or z-score < -3, will be consider as Outlier

# z-score = (data - mean) / SD
```

```python
# Finding the z-score based on age column

df3 = df.copy()

df3['z-score'] = (df3.age - df3.age.mean()) / df3.age.std()

# find the outlier
df3[(df3["z-score"] < -3) | (df3["z-score"] > 3)]

# delete these outliers
df3 = df3[(df3["z-score"] >= -3) & (df3["z-score"] <= 3)]

df3.shape
```

```python
# find the outlier
df3[(df3["z-score"] < -3) | (df3["z-score"] > 3)]
```

```python
df3 = df.copy()

df3['z-score'] = (df3.age - df3.age.mean()) / df3.age.std()

# find the outlier
df3[(df3["z-score"] < -3) | (df3["z-score"] > 3)]
```

```python
# Replace outliers with mean value

mean_value = int(df3.age.mean())
print(mean_value)
```

```python
df3.loc[(df3["z-score"] < -3) | (df3["z-score"] > 3), 'age'] = mean_value
```
